```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
train.head()
```

```python
train.loc[train['y'] > 250, 'X314']
```

```python
test[[3913, 3718, 3281, 2914], 'X0'] = 'q'
test[[311,153], 'X0'] = 'aa'

test.loc[[4187, 4177, 4161], 'X2'] = 'ar'
test.loc[[3911], 'X2'] = 'o'
test.loc[[3323], 'X2'] = 'c'
test.loc[[3073, 2930], 'X2'] = 'aa'
test.loc[[2473], 'X2'] = 'l'

test.loc[test['X5'].isin(['a', 'b', 't', 'z']), 'X5'] = 'u'


```

```python
column = 'X0'
train_median = train.groupby(column, as_index=False)['y'].median()
train_mean = train.groupby(column, as_index=False)['y'].mean()
train_group = train.groupby(column, as_index=False)['ID'].count()
test_group = test.groupby(column, as_index=False)['ID'].count()
compare_group = train_group.join(test_group.set_index(column), on=column, rsuffix="_test", how="outer")
compare_group = compare_group.join(train_median.set_index(column), on=column, rsuffix="_median", how="outer")
compare_group = compare_group.join(train_mean.set_index(column), on=column, rsuffix="_mean", how="outer")
compare_group.columns = [column, 'count_train', 'count_test', 'y_median', 'y_mean']
compare_group
```
